```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('KaggleDataset.csv', index_col='customerID')

#Converting Yes/No columns into 0/1 for SQL
binary_cols = ['Partner','Dependents','PhoneService','PaperlessBilling','Churn']
df1[binary_cols] = df1[binary_cols].apply(lambda col: col.map({'Yes': 1, 'No': 0}))

#Converting total charges to a Float
df1['TotalCharges'] = df1['TotalCharges'].replace(' ', np.nan)
df1['TotalCharges'] = df1['TotalCharges'].fillna(0)  
df1['TotalCharges'] = df1['TotalCharges'].astype('float64')

df1['gender'] = df1['gender'].map({'Male': 1, 'Female': 0})

#Establishing Groups for the Tenure Col. 
def tenureGroup(tenure):
    if tenure <= 12:
        return '0-12 months'
    elif tenure <= 24:
        return '13-24 months'
    elif tenure <= 48:
        return '25-48 months'
    else:
        return '49+ months'
df1['TenureGroup'] = df1['tenure'].apply(tenureGroup)

#Establishing a new column that shows how much Revenue is at Risk(Based off INDV Monthly Charges * Churn)
df1['RevenueAtRisk'] = df1['MonthlyCharges'] * df1['Churn']

df1.to_csv('Telco_Churn_Cleaned.csv', index=True) 
```
